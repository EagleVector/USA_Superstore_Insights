Q16 Try to find out 10 top  product with highest margins.

```python
store.head()
```

```python
store.groupby('Product Name')['Product Base Margin'].mean()
```

```python
df = pd.DataFrame(store.groupby('Product Name')['Product Base Margin'].mean())
```

Converting margins into percentage

```python
df['Product Base Margin'] = df['Product Base Margin']* 100
```

```python
df['Rank'] = df['Product Base Margin'].rank(method = 'max', ascending = False)
```

```python
df[df['Rank'] <= 10]
```
